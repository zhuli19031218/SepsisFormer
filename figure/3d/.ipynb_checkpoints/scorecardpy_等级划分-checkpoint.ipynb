```python
import pandas as pd
datascore=pd.read_csv('data/data_cq.csv')
datascore['score']=0
datascore
```

```python
for i in range(datascore.shape[0]):
    score=0.0
    if datascore['APTT'][i]<24.0:
        score=score+0.0
    elif 24.0<=datascore['APTT'][i]<37.0:
        score=score+1.0
    elif 37.0<=datascore['APTT'][i]:
        score=score+2.0

    if datascore['INR'][i]<=1.2:
        score=score+0.0
    elif 1.2<datascore['INR'][i]<=1.4:
        score=score+1.0
    elif 1.4<datascore['INR'][i]<=1.6:
        score=score+2.0
    elif 1.6<datascore['INR'][i]:
        score=score+3.0

    if datascore['age'][i]<46.0:
        score=score-1.0
    elif 46.0<=datascore['age'][i]<58.0:
        score=score+1.0
    elif 58.0<=datascore['age'][i]<83.0:
        score=score+2.0
    elif 83.0<=datascore['age'][i]:
        score=score+3.0

    if datascore['Lymphocytes'][i]< 1.1:
        score=score+2.0
    # elif 0.7<=datascore['Lymphocytes'][i]<1.1:
    #     score=score+1.0
    elif 1.1<=datascore['Lymphocytes'][i]:
        score=score+1.0

    if datascore['Monocytes'][i]<0.1:
        score=score+2.0
    elif 0.1<=datascore['Monocytes'][i]<0.6:
        score=score+1.0
    elif 0.6<=datascore['Monocytes'][i]:
        score=score+2.0

    if datascore['Neutrophil'][i]<1.8:
        score=score+2.0
    elif 1.8<=datascore['Neutrophil'][i]<6.3:
        score=score+1.0
    elif 6.3<=datascore['Neutrophil'][i]<25.0:
        score=score+2.0
    elif 25.0<=datascore['Neutrophil'][i]:
        score=score+3.0

    if datascore['White_Blood_Cells'][i]< 4.0:
        score=score+2.0
    elif 4.0<=datascore['White_Blood_Cells'][i]< 10.0:
        score=score+1.0
    elif 10<=datascore['White_Blood_Cells'][i]<28.0:
        # elif 10<=datascore['Neutrophils'][i]<28.0:
        score=score+2.0
    elif 28.0<=datascore['White_Blood_Cells'][i]:
        score=score+3.0

    if datascore['Platelet_Count'][i]<125.0:
        score=score+2.0
    # elif 125<=datascore['PlateletCount'][i]<350.0:
    #     score=score+1.0
    elif 125.0<=datascore['Platelet_Count'][i]:
        score=score+1.0
    datascore['score'][i]=round(score)
datascore
```

```python
# 9,14,18
#9,14,17
data_score=datascore
data_score['hierarchy']=0
c1=c2=c3=c4=0
d1=d2=d3=d4=0
for i in range(data_score.shape[0]):    
    if data_score['score'][i]<9.0:
        if data_score['dead'][i]==1:
            c1=c1+1
        d1=d1+1
        data_score['hierarchy'][i]=1
    elif 9.0<=data_score['score'][i]<13.0:
        if data_score['dead'][i]==1:
            c2=c2+1
        d2=d2+1
        data_score['hierarchy'][i]=2
    elif 13.0<=data_score['score'][i]<17.0:
        if data_score['dead'][i]==1:
            c3=c3+1
        d3=d3+1
        data_score['hierarchy'][i]=3
    elif 17.0<=data_score['score'][i]:
        if data_score['dead'][i]==1:
            c4=c4+1
        d4=d4+1
        data_score['hierarchy'][i]=4
print(c1,c2,c3,c4,"\n",d1,d2,d3,d4)
print(c1/d1,c2/d2,c3/d3,c4/d4)
data_score.to_csv('aaa.csv')
```
