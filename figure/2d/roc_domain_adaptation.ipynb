```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def del_duplicate(x_list, y_list):
    real_inx = -1
    x_new = []
    y_new = []
    for inx, x in enumerate(x_list):
        if x not in x_new:
            x_new.append(x)
            real_inx += 1
            # print(real_inx, inx)
            y_new.append([y_list[inx]])
        else:
            y_new[real_inx].append(y_list[inx])
    # 对y_new取平均 sum(y)/len(y)
    y_new = [max(y) for y in y_new]
    # 第一个数必须为0
    # return np.array(x_new), np.array(y_new)
    return x_new, y_new
def ban_down(y_list):
    old_one = 0
    y_new = []
    for inx, y in enumerate(y_list):
        if old_one < y <= 1:
            old_one = y
        y_new.append(old_one)  # if old_one <= 1 else 1
    return y_new
```

```python
local_real_lable=pd.read_csv(r"logs\domain_adaptation\zt\test_labels_no_domain.csv",header=None).to_numpy()[:, -1]

local_predict_lable=pd.read_csv(r"logs\domain_adaptation\zt\pre_test_result_no_domain.csv",header=None).to_numpy()[:, -1]
# print(mean_teacher_predict_lable.shape,mean_teacher_real_lable.shape)
# 计算真阳性率和假阳性率
fpr_local, tpr_local, thresholds_local = roc_curve(local_real_lable, local_predict_lable)
roc_auc_local = auc(fpr_local, tpr_local)
fpr_local, tpr_local = del_duplicate(fpr_local, tpr_local)
z6_local= np.polyfit(fpr_local, tpr_local, 10)
p6_local = np.poly1d(z6_local)
y_pred_local = p6_local(fpr_local)
print('Local test auc:',roc_auc_local)
```
